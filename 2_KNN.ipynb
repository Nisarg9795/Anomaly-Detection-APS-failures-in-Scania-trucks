```python
import pandas as pd 
from scipy import stats
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from scipy.spatial import distance
```

```python
df = pd.read_csv("aps_failure_training_set.csv") 
df_train = pd.read_csv("aps_failure_test_set.csv")
#print(df.shape)


df = df.rename(columns = {'class' : 'Flag'})
df['Flag'] = df.Flag.map({'neg':0, 'pos':1})
df = df.replace(['na'],[np.NaN])
#print(df.head())


df_X= df.loc[:,df.columns != 'Flag']
df_Y= df.loc[:,df.columns == 'Flag']
df_X=df_X.apply(pd.to_numeric)
df_Y=df_Y.apply(pd.to_numeric)


df_train = df_train.rename(columns = {'class' : 'Flag'})
df_train['Flag'] = df_train.Flag.map({'neg':0, 'pos':1})
df_train = df_train.replace(['na'],[np.NaN])


df_train_X= df_train.loc[:,df.columns != 'Flag']
df_train_Y= df_train.loc[:,df.columns == 'Flag']
#print(df_train_Y)
df_train_X=df_train_X.apply(pd.to_numeric)
df_train_Y=df_train_Y.apply(pd.to_numeric)
```

```python
from sklearn.neighbors import KNeighborsClassifier
## Instantiate the model with 5 neighbors. 
knn = KNeighborsClassifier(n_neighbors=150)
## Fit the model on the training data.
knn.fit(df_X, df_Y.values.ravel())
## See how the model performs on the test data.
Y_pred=knn.predict(df_train_X)
print(confusion_matrix(df_train_Y.values.ravel(),Y_pred))



#print(df_X.head())
#df_X[(np.abs(stats.zscore(df_X)) < 3).all(axis=1)]
#print(df_X)
#print(df_X["aa_000"]-df_X["aa_000"].mean())
#print(df_X.ix[:,0]-df_X.ix[:,0].mean())
#print(df_X["aa_000"])

#df_X=df_X[np.abs(df_X.aa_000-df_X.aa_000.mean()) <= (3*df_X.aa_000.std())]
#df_X=df_X[np.abs(df_X.ab_000-df_X.ab_000.mean()) <= (3*df_X.ab_000.std())]
#df_X=df_X[np.abs(df_X.ac_000-df_X.ac_000.mean()) <= (3*df_X.ac_000.std())]
```

```python
distances=(distance.cdist(df_train_X,df_X))
print(np.shape(distances))
```

```python

predicted_label = []
for i in range (0,16000) :
    sorted_distances=np.argsort(distances[i,:])
    v=sorted_distances[:500]
    temp_output = []
    for index in v:
        if df_Y.iloc[index]["Flag"] == 1:
            temp_output.append(1)
            if len(temp_output) == 3:
                break
    if len(temp_output) == 3:
        predicted_label.append(1)
    else:
        predicted_label.append(0)

print(confusion_matrix(df_train_Y.values.ravel(),predicted_label))   
    
```
